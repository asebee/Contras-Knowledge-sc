```python
import numpy as np
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt

# 设置随机种子以确保结果可重复
np.random.seed(42)

# 定义聚类的数量和每个聚类的样本数量
n_clusters = 4
n_samples_per_cluster = 200

# 生成多个2D高斯分布的聚类数据
centers = [[-5, -5], [0, 0], [5, 5], [10, -5]]  # 聚类中心
cluster_std = [1.0, 1.5, 1.0, 2.0]  # 每个聚类的标准差

X, y = make_blobs(n_samples=n_clusters * n_samples_per_cluster,
                  centers=centers,
                  cluster_std=cluster_std,
                  random_state=42)

# 绘制数据点
plt.figure(figsize=(8, 8))
for i in range(n_clusters):
    plt.scatter(X[y == i, 0], X[y == i, 1], label=f'Cluster {i+1}', alpha=0.6)

# 添加图例和标题
plt.legend()
plt.title('2D Gaussian Clusters')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
#计算两列数据相关系数的函数
import numpy as np
def calculate_correlation_coefficient(data1, data2):
    """
    计算两列数据的相关系数
    :param data1: 第一列数据
    :param data2: 第二列数据
    :return: 相关系数
    """
    # 确保数据长度相同
    if len(data1) != len(data2):
        raise ValueError("两列数据长度不一致")

    # 计算相关系数
    correlation_matrix = np.corrcoef(data1, data2)
    return correlation_matrix[0, 1]
```
