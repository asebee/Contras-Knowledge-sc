```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
training_set_proportions = [0.1, 0.3, 0.5, 0.7, 1.0]
accuracy_scanpy = [0.535, 0.546, 0.553, 0.551, 0.547]  # Scanpy (2017)
accuracy_scbert = [0.629, 0.67, 0.691, 0.717, 0.73]  # scBERT (2022)
accuracy_scfoundation = [0.658, 0.689, 0.72, 0.729, 0.746]  # scFoundation (2024)
accuracy_Knowlege_sc = [0.679, 0.713, 0.728, 0.733, 0.742]  # Label-sc
```

```python
plt.rcParams['font.family'] = 'SimHei'  # 使用黑体字
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
plt.figure(figsize=(4, 3))

# Plot each method's accuracy
plt.plot(training_set_proportions, accuracy_scanpy, marker='o', linestyle='--',label='Scanpy (2017)')
plt.plot(training_set_proportions, accuracy_scbert, marker='s', linestyle=':', label='scBERT (2022)')
plt.plot(training_set_proportions, accuracy_scfoundation, marker='^',linestyle='-.', label='scFoundation (2024)')
plt.plot(training_set_proportions, accuracy_Knowlege_sc, marker='d', linestyle='-',label='Knowledge-sc')

# Add title and labels
#plt.title('Accuracy Changes of Different Methods with Training Set Proportions', fontsize=14)
plt.xlabel('训练集比例', fontsize=12)
plt.ylabel('Macro-F1', fontsize=12)

# Add legend
plt.legend(fontsize=9, loc='best')

# Add grid
plt.grid(alpha=0.3)

# Show plot
plt.tight_layout()
plt.savefig('annotation_data.svg', format='svg', bbox_inches='tight')
plt.show()
```

```python
lambda_values = [0, 0.2, 0.5, 0.8, 1.0]
pcc_scores = [None, 0.639, 0.642, 0.645, 0.649]
nmi_scores = [None, 0.734, 0.748, 0.744, 0.746]
macro_f1_scores = [0.738, 0.741, 0.742, 0.739, 0.736]
# 设置中文字体
plt.rcParams['font.family'] = 'SimHei'  # 使用黑体字
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 创建图形
plt.figure(figsize=(4, 3))

# 绘制三条折线，使用不同的线型和标记
plt.plot(lambda_values, pcc_scores, marker='o', linestyle='-', linewidth=2, label='PCC')
plt.plot(lambda_values, nmi_scores, marker='s', linestyle='--', linewidth=2, label='NMI')
plt.plot(lambda_values, macro_f1_scores, marker='^', linestyle='-.', linewidth=2, label='Macro-F1')

# 设置x轴标签位置
plt.xticks(lambda_values)

# 添加标题和标签
# plt.title('Lambda参数对评估指标的影响', fontsize=14)
plt.xlabel('$\lambda$ 值（重建损失比例）', fontsize=12)
plt.ylabel('指标值', fontsize=12)

# 添加图例
plt.legend(fontsize=12, loc='best')

# 添加网格线
plt.grid(alpha=0.3)

# 显示图形
plt.savefig('parameter_analysis_2.svg', dpi=300, bbox_inches='tight')
plt.show()
```
