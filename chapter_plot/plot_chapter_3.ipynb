```python
import numpy as np
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt

# 
np.random.seed(42)

# 
n_clusters = 4
n_samples_per_cluster = 200

# 2D
centers = [[-5, -5], [0, 0], [5, 5], [10, -5]]  # 
cluster_std = [1.0, 1.5, 1.0, 2.0]  # 

X, y = make_blobs(n_samples=n_clusters * n_samples_per_cluster,
                  centers=centers,
                  cluster_std=cluster_std,
                  random_state=42)

# 
plt.figure(figsize=(8, 8))
for i in range(n_clusters):
    plt.scatter(X[y == i, 0], X[y == i, 1], label=f'Cluster {i+1}', alpha=0.6)

# 
plt.legend()
plt.title('2D Gaussian Clusters')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
#
import numpy as np
def calculate_correlation_coefficient(data1, data2):
    """
    
    :param data1: 
    :param data2: 
    :return: 
    """
    # 
    if len(data1) != len(data2):
        raise ValueError("")

    # 
    correlation_matrix = np.corrcoef(data1, data2)
    return correlation_matrix[0, 1]
```
